## Basics
A class is a collection of objects. Classes are blueprints for creating objects. A class defines a set of attributes and methods that the created objects (instances) can have.  


Key ideas:

- Classes are created by keyword class.
- Attributes are the variables that belong to a class
- Attributes are always public and can be accessed using the dot (.) operator.

#### Dynamic Typing: 
Python classes and objects don’t require explicit type declarations.

````python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canine"  

    def __init__(self, name, age):
        # Instance attributes (unique to each object)
        self.name = name 
        self.age = age  
    
    # Instance method
    # self represents the instance of the class. It must be the first parameter of instance methods.
    # Python automatically passes the instance as the first argument, but Python does not enforce the name itself; 
    # you could technically use another name, though that is bad practice.
    def bark(self):
        return f"{self.name} says Woof!"
````

## Objects

An object is an instance of a class. It has:

- State: It is represented by the attributes and reflects the properties of an object.
- Behavior: It is represented by the methods of an object and reflects the response of an object to other objects.
- Identity: It gives a unique name to an object and enables one object to interact with other objects.

````python
# Creating an object of the Dog class
dog1 = Dog("Buddy", 3)

print(type(dog1))  # <class '__main__.Dog'>

# ways to call method
print(Dog.bark(dog1))  # via class
print(dog1.bark())     # via objects
````

## Dynamic Methods

In Python, methods can be added dynamically to a class or to a single object at runtime.

- Adding a method to a class makes it available to current and future instances.
- Adding a method to a single object affects only that object.

This flexibility is one reason Python is considered highly dynamic.

## Deletion of Attributes

The `del` statement removes a name or attribute binding. It does not directly destroy an object unless that was the last reference.

For classes:

- `del obj.attr` deletes an instance attribute from a specific object.
- `del ClassName.attr` deletes a class attribute from the class itself.

````python
class Demo:
    class_attr = "I am class attribute"
    shadowed = "class version"

obj = Demo()

# instance attribute
obj.instance_attr = "I am instance attribute"

# instance attribute shadows class attribute
# If an instance attribute was shadowing a class attribute, 
# deleting the instance attribute can make the class attribute visible again.
obj.shadowed = "instance version"


print(obj.class_attr)      # from class
print(obj.instance_attr)   # from object
print(obj.shadowed)        # instance version
````

````python
# 1) Delete class attribute from object,
del obj.class_attr

# O/P - Attribute error
# Why: class_attr is not stored in obj; it belongs to the class.
# So Python cannot delete it from the object.
````

````python
# 2) Delete instance attribute from object
del obj.instance_attr
# it is deleted successfully.


print(obj.instance_attr)
# AttributeError
# Why: it existed only on that object, and now it is gone.
````

````python
# 3) Delete instance attribute that shadows class attribute

print(obj.shadowed)   # instance version
del obj.shadowed
print(obj.shadowed)   # now falls back to class version
````

## Inheritance

Inheritance allows one class to reuse and extend the behavior of another class.

- A parent or base class provides shared behavior.
- A child or derived class inherits from the parent class.
- Methods can be overridden in the child class when needed.

````python

# Parent class (Base class/Superclass)
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"
    
    def info(self):
        return f"I am {self.name}"

# Child class (Derived class/Subclass)
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)  # Call parent constructor
        self.breed = breed
    
    # Method overriding
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

dog = Dog("Buddy", "Labrador")
cat = Cat("Whiskers")

print(dog.speak())  # Buddy says Woof!
print(cat.speak())  # Whiskers says Meow!
print(dog.info())   # I am Buddy - inherited method
print(dog.breed)    # Labrador - child-specific attribute
````

## Multiple Inheritance

Python supports multiple inheritance, where a class can inherit from more than one parent class.

This is useful when you want to combine behaviors from different base classes.

````python
class Flyer:
    def fly(self):
        return "Flying high!"

class Swimmer:
    def swim(self):
        return "Swimming fast!"

class Duck(Flyer, Swimmer):
    def quack(self):
        return "Quack quack!"

duck = Duck()
print(duck.fly())    # Flying high!
print(duck.swim())   # Swimming fast!
print(duck.quack())
````

## MRO

MRO stands for Method Resolution Order. It defines the order in which Python looks for methods and attributes when multiple inheritance is involved.

You can inspect it with the `mro()` method or the `__mro__` attribute.

````python
# Base class
class A:
    label = "A: Base class"

# B inherits from A
class B(A):
    label = "B: Masala blend"

# C also inherits from A
class C(A):
    label = "C: Herbal blend"

# Multiple inheritance: D inherits from both C and B
# This creates a "diamond" pattern: D -> C -> A and D -> B -> A
class D(C, B):
    pass  # D doesn't override label, so MRO determines which one is used

# Create instance
cup = D()

# Access label - which one is used?
print(cup.label)  # Output: C: Herbal blend

# Print the Method Resolution Order
print(D.__mro__)  
# Output: (<class '__main__.D'>, <class '__main__.C'>, <class '__main__.B'>, <class '__main__.A'>, <class 'object'>)

````

#### MRO Order: D → C → B → A → object

**Why this order?**

1. **D first**: Check the class itself
2. **C before B**: Because D(C, B) lists C first (left-to-right order)
3. **B after C**: Second parent in inheritance list
4. **A after both B and C**: Common parent comes after all children
5. **object last**: All classes inherit from object in Python 3

#### C3 Linearization Rules
Python uses the C3 linearization algorithm which ensures:

1. **Children before parents**: Subclasses are checked before their parent classes
2. **Left-to-right order**: Parents listed first have priority
3. **Monotonicity**: If C precedes B in one class, C must precede B everywhere
4. **No duplicates**: Each class appears only once in the MRO

#### Best Practices

- **Prefer composition over complex inheritance**: Multiple inheritance can be confusing
- **Use super() consistently**: Always use `super()` in multiple inheritance scenarios
- **Check MRO during development**: Use `ClassName.__mro__` to verify expected behavior
- **Document inheritance**: Clearly document why multiple inheritance is needed

### Static Methods in Python

A method that belongs to a class but doesn't require instance or class data.


- **@staticmethod decorator**: Marks a method as static (no automatic `self` or `cls` parameter)
- **No instance needed**: Can be called on the class itself without creating an instance
- **Utility functions**: Often used for utility functions related to the class
- **No access to self or cls**: Cannot access or modify instance or class state

````python
class ChaiUtils:
    # @staticmethod decorator - marks this as a static method
    @staticmethod
    def clean_ingredients(text):
        # No 'self' parameter - this method doesn't need instance data
        # Pure utility function that processes input and returns result
        return [item.strip() for item in text.split(",")]


# Input string with spaces
raw = " water , milk , ginger , honey "

# Call static method directly on the class (no instance needed)
cleaned = ChaiUtils.clean_ingredients(raw)
print(cleaned)  # Output: ['water', 'milk', 'ginger', 'honey']
````

### Class Methods in Python

- **Class method**: A method that receives the class itself as the first parameter (`cls`)
- **@classmethod decorator**: Marks a method as a class method
- **Alternative constructors**: Primary use case for creating instances in different ways
- **Access to class**: Can access and modify class attributes and call other class methods
- **Factory pattern**: Often used to implement factory methods for object creation

````python
class ChaiOrder:
    def __init__(self, tea_type, sweetness, size):
        self.tea_type = tea_type
        self.sweetness = sweetness
        self.size = size

    # Class method - receives 'cls' as first parameter (the class itself)
    @classmethod
    def from_dict(cls, order_data):
        """Alternative constructor: Create ChaiOrder from dictionary"""
        # cls refers to ChaiOrder class
        return cls(
            order_data["tea_type"],
            order_data["sweetness"],
            order_data["size"],
        )

    # Another class method - different way to construct object
    @classmethod
    def from_string(cls, order_string):
        """Alternative constructor: Create ChaiOrder from string format"""
        # Parse string and create object using cls()
        tea_type, sweetness, size = order_string.split("-")
        return cls(tea_type, sweetness, size)

class ChaiUtils:
    # Static method for comparison - no cls or self parameter
    @staticmethod
    def is_valid_size(size):
        """Utility function - doesn't need class or instance data"""
        return size in ["Small", "Medium", "Large"]


# Static method example - call directly on class
print(ChaiUtils.is_valid_size("Medium"))  # Output: True

# Class method example 1: Create from dictionary
order1 = ChaiOrder.from_dict({"tea_type": "masala", "sweetness": "medium", "size":"Large"})

# Class method example 2: Create from string
order2 = ChaiOrder.from_string("Ginger-Low-Small")

# Regular constructor - direct instantiation
order3 = ChaiOrder("Large", "Low", "Large")

# Display instance attributes
print(order1.__dict__)  # Output: {'tea_type': 'masala', 'sweetness': 'medium', 'size': 'Large'}
print(order2.__dict__)  # Output: {'tea_type': 'Ginger', 'sweetness': 'Low', 'size': 'Small'}
print(order3.__dict__)  # Output: {'tea_type': 'Large', 'sweetness': 'Low', 'size': 'Large'}
````

#### Comparison: @classmethod vs @staticmethod

| Feature                    | @classmethod                              | @staticmethod                            |
| -------------------------- | ----------------------------------------- | ---------------------------------------- |
| **First parameter**        | Receives `cls` (the class itself)         | Receives **no** automatic first argument |
| **Use case**               | Operate on the **class**, not instance    | Utility function related to the class    |
| **Access to cls**          | ✅ Yes                                    | ❌ No                                    |
| **Access to self**         | ❌ No                                     | ❌ No                                    |
| **Can create instances**   | ✅ Yes (using `cls()`)                    | ⚠️ Yes (but must hardcode class name)    |
| **Works with inheritance** | ✅ Yes (cls changes with subclass)        | ⚠️ Limited (hardcoded class name)        |
| **Common use**             | Alternative constructors, factory methods | Utility/helper functions                 |

#### Usecases of Classmethod

1. **Alternative Constructors** (Most Common)

   ```python
   @classmethod
   def from_json(cls, json_string):
       data = json.loads(json_string)
       return cls(data['field1'], data['field2'])
   ```

2. **Factory Methods**

   ```python
   @classmethod
   def create_default(cls):
       return cls("Default Tea", "Medium", "Large")
   ```

3. **Accessing/Modifying Class Attributes**

   ```python
   order_count = 0

   @classmethod
   def increment_orders(cls):
       cls.order_count += 1
   ```

4. **Inheritance-Friendly Object Creation**

   ```python
   class SpecialChai(ChaiOrder):
       pass

   # from_string returns SpecialChai, not ChaiOrder
   special = SpecialChai.from_string("Masala-High-Large")
   ```

### Property Decorators in Python

- **@property decorator**: Turns a method into a read-only attribute (getter)
- **@property_name.setter**: Defines a setter method for the property
- **Encapsulation**: Hide internal representation and control access to attributes
- **Computed properties**: Properties that calculate values dynamically
- **Validation**: Enforce rules when setting attribute values
- **Private attributes**: Convention to use underscore prefix (`_attribute`) for internal attributes
- Used in **Encapsulation**

````python
class TeaLeaf:
    def __init__(self, age):
        # Private attribute (by convention) - use underscore prefix
        self._age = age

    # @property decorator - creates a getter method
    # Now 'age' can be accessed like an attribute: leaf.age
    @property
    def age(self):
        """Getter: Returns computed value (adds 2 years)"""
        return self._age + 2

    # @age.setter - creates a setter method for 'age' property
    # Now 'age' can be set like an attribute: leaf.age = value
    @age.setter
    def age(self, age):
        """Setter: Validates and sets the age with constraints"""
        if 1 <= age <= 5:
            self._age = age  # Store in private attribute
        else:
            # Validation - raise error for invalid values
            raise ValueError("Tea leaf age must be between 1 and 5 years")


# Create instance with age 2
leaf = TeaLeaf(2)

# Access property like an attribute (calls the getter)
print(leaf.age)  # Output: 4 (2 + 2 from getter logic)

# Set property like an attribute (calls the setter with validation)
leaf.age = 6  # This will raise ValueError because 6 > 5

# This line won't execute because of the error above
print(leaf.age)
````

````python
class ChaiCup:
    def __init__(self, volume_ml, temperature):
        self._volume_ml = volume_ml
        self._temperature = temperature

    # Read-only property (no setter)
    @property
    def volume_oz(self):
        """Convert milliliters to ounces"""
        return self._volume_ml * 0.033814

    # Read-write property with validation
    @property
    def temperature(self):
        """Get temperature"""
        return self._temperature

    @temperature.setter
    def temperature(self, value):
        """Set temperature with validation"""
        if not 60 <= value <= 100:
            raise ValueError("Temperature must be between 60-100°C")
        self._temperature = value

    # Computed property
    @property
    def is_hot(self):
        """Determine if chai is hot based on temperature"""
        return self._temperature > 70


cup = ChaiCup(250, 85)
print(cup.volume_oz)  # 8.4535 (computed from ml)
print(cup.temperature)  # 85
print(cup.is_hot)  # True

cup.temperature = 65  # Valid - setter allows this
# cup.volume_oz = 10  # Error! No setter defined (read-only)
# cup.temperature = 110  # Error! Validation fails
````

````python
# Without @property (Direct Attribute Access)
class TeaLeaf:
    def __init__(self, age):
        self.age = age  # Direct attribute - no validation

leaf = TeaLeaf(100)  # Can set any value - no validation!
print(leaf.age)  # Direct access
leaf.age = -50  # Can set invalid value - no control
````

````python
# With @property (Controlled Access)
class TeaLeaf:
    def __init__(self, age):
        self._age = age  # Private attribute

    @property
    def age(self):
        return self._age  # Controlled read access

    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value  # Controlled write access with validation

leaf = TeaLeaf(2)
print(leaf.age)  # Looks like attribute access, actually calls getter
leaf.age = -5  # ValueError - validation prevents invalid data
````

#### Private Attribute Convention

```python
self._age  # Private by convention (single underscore)
```

- Single underscore (`_attribute`) indicates "internal use only"
- Not actually private (Python has no true private attributes)
- Convention tells other developers not to access directly
- The property provides the public interface

### Composition
 A class contains instances of other classes (HAS-A relationship)
 Use when class needs functionality from another class ("has-a")

#### Note

Use inheritance for genuine "is-a" relationships that rarely change (e.g., Dog is-a Animal). Use composition when you're really modeling "has-a" / "uses-a" relationships, or when behavior needs to be flexible, swappable, or mixed-and-matched.

**A simple rule of thumb** - 
If you find yourself asking "does this really behave exactly like its parent, with a few tweaks?" → inheritance might fit.
If you're asking "does this object use another object to get its job done?" → that's composition, and it's usually the safer, more maintainable choice.

````python
class Engine:
    def start(self):
        print("Engine starting...")
    
    def stop(self):
        print("Engine stopping...")

class GPS:
    def navigate(self, destination):
        print(f"Navigating to {destination}")

class Car:
    def __init__(self):
        self.engine = Engine()   # composed of an Engine object
        self.gps = GPS()         # composed of a GPS object

    def start(self):
        self.engine.start()      # delegate work to the component

    def navigate(self, destination):
        self.gps.navigate(destination)

car = Car()
car.start()                  # Engine starting...
car.navigate("New York")     # Navigating to New York
````
