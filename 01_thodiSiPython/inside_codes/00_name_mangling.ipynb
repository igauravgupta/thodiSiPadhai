### Name Mangling in Python
Name mangling is a mechanism where Python automatically renames attributes that start with double underscores (__) inside a class, making them harder to access from outside.

#### How it works
When you write `__attribute` inside a class, Python renames it to `_ClassName__attribute` behind the scenes.

```python
class MyClass:
    def __init__(self):
        self.public = "I'm public"
        self._protected = "I'm protected (convention)"
        self.__private = "I'm private (mangled)"

obj = MyClass()

print(obj.public)          # ✅ Works fine
print(obj._protected)      # ✅ Works (but discouraged)
print(obj.__private)       # ❌ AttributeError!
print(obj._MyClass__private)  # ✅ Works — mangled name
```

#### Why Python does this
1. Prevent accidental overriding in subclasses
This is the primary reason. If a parent class uses __attr, a subclass with the same attribute name won't accidentally clobber it:


```python
class Base:
    def __init__(self):
        self.__value = 10  # becomes _Base__value

class Child(Base):
    def __init__(self):
        super().__init__()
        self.__value = 99  # becomes _Child__value — different!

c = Child()
print(c._Base__value)   # 10 — untouched!
print(c._Child__value)  # 99
# Without mangling, Child would silently overwrite Base's __value.
```

2. Signal strong "internal use only" intent
It communicates to other developers: "don't touch this directly."

#### Naming convention summary

| Syntax | Meaning | Accessible from outside? |
|---|---|---|
| `attr` | Public | Yes |
| `_attr` | Protected (convention) | Yes, but discouraged |
| `__attr` | Private (mangled) | Only via `_ClassName__attr` |
| `__attr__` | Dunder / magic | Yes (used by Python itself) |



#### Note: 
`__attr__` (double underscore on both sides) is not mangled — those are Python's special/magic methods like `__init__`, `__str__`, etc.

#### Key things to remember

1. Name mangling is not true privacy — you can still access it via the mangled name. Python follows the principle "we're all consenting adults here."
2. It's not about security, it's about avoiding naming collisions in inheritance hierarchies.
3. Use __ only when you genuinely need to protect against subclass overrides — for general "private" intent, _single_underscore is the more common convention.
